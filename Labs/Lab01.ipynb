## Project 1: Exploratory Data Analysis - Gapminder dataset

The Gapminder dataset provides population data from 1952 to 2007 (at 5 year Intervals) for several countries around the world.Compares the populations of the European countries France, United Kingdom, Italy, Germany and Spain over this period using a line chart. Make appropriate modifications to the chart title, axis titles, legends, figure size, font size, colors etc to make the chart readable and visually appealing - #Line Chart  

```python
import pandas as pd
import numpy as np
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python

plt.rcParams['font.size'] = 14
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.facecolor'] = '#00000000'
```

```python
df_gapminder = pd.read_csv(r"C:\Users\gaura\OneDrive\Documents\GitHub\MachineLearning2026_Gaurav\Dataset\gapminderDataFiveYear.csv")
df_gapminder.columns
```

```python
countries = ["France", "United Kingdom", "Italy", "Germany", "Spain"]
eu_df = df_gapminder[df_gapminder["country"].isin(countries)]

sns.set_style("darkgrid")
plt.figure(figsize=(12, 6))

sns.lineplot(
    data=eu_df,
    x="year",
    y="pop",
    hue="country",
    marker="o",
    linewidth=2
)

plt.title("Population Comparison of Major European Countries (1952–2007)", fontsize=16)
plt.xlabel("Year", fontsize=14)
plt.ylabel("Population", fontsize=14)
plt.legend(title="Country", fontsize=12)

plt.tight_layout()
plt.show()
```
