Now we will using the formula 

m = Σ((x - mean_x)(y - mean_y)) / Σ((x - mean_x)^2)

for slope

b = mean_y − m * mean_x

where, 

X = input feature

Y = target

```python
mean_x = np.mean(X)
mean_y = np.mean(Y)

m = np.sum((X - mean_x)*(Y - mean_y)) / np.sum((X - mean_x)**2)
b = mean_y - m*mean_x
```

```python
Y_pred = m*X + b
```
